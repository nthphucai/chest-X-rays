```python
import pandas as pd 
import numpy as np
import os 
import sys
```

```python
classes = np.load('/Users/HPhuc/Practice/12. classification/vinbigdata/output/columns_14.npy')
classes 
```

```python
data = pd.read_csv('/Users/HPhuc/Practice/12. classification/vinbigdata/output/csv/data.csv', index_col=0)
data
```

```python
counts = pd.DataFrame([data[c].value_counts() for c in classes])
counts 
```

```python
N = counts.iloc[0, 0] + counts.iloc[0, 1] 
beta = (N - 1) / N
weights = (1 - beta) / ( 1 - (beta ** counts))
weights 
```

```python
normalized_weight = weights / weights.values.sum(axis=1)[:, np.newaxis] * 2 
normalized_weight = normalized_weight / normalized_weight.values[:, 0, np.newaxis]
normalized_weight
```
