```python
import os 
import pandas as pd 
import numpy as np 
import cv2
```

```python
train_df = pd.read_csv('/Users/HPhuc/Practice/12. classification/vinbigdata/output/csv/total_data.csv', index_col=0)
train_df = train_df[['image_id', 'class_name']]
train_df
```

```python
ids_ = train_df['image_id'].unique()
print(len(ids_))
data = pd.DataFrame(columns=['image_ids', 'class_name'])
data['image_ids'] = ids_
data
```

```python
for idc, id_ in enumerate(ids_):
    data.loc[idc,'class_name'] = train_df[train_df.image_id == str(id_)].class_name.values
    
data 
```

```python
classes = ['Aortic enlargement', 'Atelectasis', 'Calcification', 'Cardiomegaly', 'Consolidation', 'ILD',
              'Infiltration', 'Lung Opacity', 'Nodule/Mass', 'Other lesion', 'Pleural effusion', 'Pleural thickening',
              'Pneumothorax', 'Pulmonary fibrosis']

for idc in range(0, len(data)):
    for c in classes:
        data.loc[idc, c] = 1 if c in data.loc[idc, 'class_name'] else 0

data 
```

```python
train_path = '/Users/HPhuc/Practice/12. classification/vinbigdata/data/train/'
data['image_paths'] = f'{train_path}' + data['image_ids'] + '.png'
data 
```

```python
data = data.rename(columns={'Nodule/Mass': 'Nodule Mass'})

import parameter as para 
data.to_csv(os.path.join(para.csv_path, 'data_train.csv')) 
data        
```
