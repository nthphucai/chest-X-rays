```python
import os
import glob

import cv2
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import glob
import os 

data_path = "/content/drive/MyDrive/Classification2D/vinbigdata/data/train"
lung_segmentation_path = "/content/drive/MyDrive/Classification2D/chest-X-rays/output/vin-lung-seg/"

img_lst= glob.glob(f"{lung_segmentation_path}/*")
idc = np.random.choice(len(img_lst))

len(img_lst)
```

```python
id_image = os.path.basename(img_lst[idc]).split(".png")[0]
```

```python
org_img_path = f"{data_path}/{id_image}.png"
print(org_img_path)

org_img = cv2.imread(org_img_path).transpose(-1, 0, 1)
org_img.shape
```

```python
crop_img_path = img_lst[idc]
print(crop_img_path)
crop_img = np.load(crop_img_path)

crop_img.shape
```

```python
_, axes = plt.subplots(1, 2, figsize=(12, 8))

axes[0].imshow(org_img[0], cmap=plt.cm.gray)
axes[1].imshow(crop_img, cmap=plt.cm.gray)
plt.show()
```
