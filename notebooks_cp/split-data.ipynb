```python
import os
import glob 

files = [os.path.basename(file_path).split(".")[0] + ".png" for file_path in glob.glob("/content/drive/MyDrive/Classification2D/chest-X-rays/output/lungs-segmentation/*")]
```

```python
train_remain = data[data["image_id"].isin(files)]
train_remain.head()
```

```python
train_remain["cropped_img_path"] = "/content/drive/MyDrive/Classification2D/chest-X-rays/output/lungs-segmentation/" + train_remain["image_id"] + ".npy"
```

```python
train_remain["cropped_img_path"] = train_remain["cropped_img_path"].apply(lambda row: row.split(".png")[0] + ".npy")
```

```python
train_remain.loc[0, "cropped_img_path"]
```

```python
def read_text_file(file_path: str):
    with open(file_path, "r") as f:
        context = [item.strip() for item in f.readlines()]
    return context

test_data = read_text_file("/content/drive/MyDrive/Classification2D/NIH-Chest-X-ray-Dataset/test_list.txt")
print(len(test_data))
train_val_data = read_text_file("/content/drive/MyDrive/Classification2D/NIH-Chest-X-ray-Dataset/train_val_list.txt")
print(len(train_val_data))
```

Train Val Data

```python
train_val_df = train_remain[~train_remain["image_id"].isin(list(test_data))]
len(train_val_df)
```

```python
train_val_df.to_csv("/content/drive/MyDrive/Classification2D/NIH-Chest-X-ray-Dataset/train_val_data_25000.csv")
```

Test Data

```python
test_df = train_remain[train_remain["image_id"].isin(list(test_data))]
len(test_df)
```
